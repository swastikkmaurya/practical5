```python
'''4. A dictionary stores student names and marks .
write a python  progam to find the student who scored the highest'''

marks = {"A":85,"B":79,"C":90}
topper =max(marks,key=marks.get)
print("topper : ",topper)
print("Marks: ", marks[topper])
```
